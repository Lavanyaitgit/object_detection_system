```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('cat.jpg')
```

```python
cv2_imshow(img)
```

```python
classNames = []
classFile = 'coco.names'
with open(classFile, 'rt') as f:
  classNames = f.read().rstrip('\n').split('\n')
```

```python
print(classNames)
```

```python
configPath = '/content/yolov3.cfg'
weightsPath = '/content/yolov3.weights'
```

```python
net = cv2.dnn_DetectionModel (weightsPath, configPath)
net.setInputSize(320, 320)
net.setInputScale((127.5, 127.5, 127.5))
net.setInputSwapRB(True)
```

```python
classIds, confs, bbox = net.detect(img, confThreshold=0.7)
```

```python
print(classIds, bbox)
```

```python
for classId, confidence, box in zip(classIds, confs, bbox):
    # Check if any objects were detected
    if classId:
        x, y, w, h = box
        cv2.rectangle(img, classNames[classId-1], (x+10, y+30), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 255, 0), 2)
```

```python
!pip install torch torchvision torchaudio
!pip install opencv-python

import torch
import cv2
from google.colab.patches import cv2_imshow

# Load the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Load the image
img = cv2.imread('cat.jpg')

# Perform object detection
results = model(img)

# Check if a cat is detected
for *xyxy, conf, cls in results.xyxy[0]:
  if model.names[int(cls)] == 'cat':
    print("Cat detected!")
    # Add a bounding box and label to the image
    x, y, w, h = map(int, xyxy)
    cv2.rectangle(img, (x, y), (w, h), (0, 255, 0), 2)
    cv2.putText(img, 'Cat', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# Display the image with bounding boxes and labels (if any)
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
